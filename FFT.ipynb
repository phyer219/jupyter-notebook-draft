```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack
N = 600
L = 5
x = np.linspace(0,L,N)
f = np.sin(66*x)+np.sin(89*x)+np.sin(197*x)+np.sin(286*x)
#f = np.exp(-100*x**2)
g = np.fft.fft(f)
g = scipy.fftpack.fft(f)
xf = np.linspace(0,2*np.pi*N/(2*L),N/2)
#print(g.real)
plt.plot(xf,2/N*np.abs(g[:N//2]))
plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack

# Number of samplepoints
N = 600
# sample spacing
T = 1.0 / 800.0
x = np.linspace(0.0, N*T, N)
y = np.sin(50.0 * 2.0*np.pi*x) + 0.5*np.sin(80.0 * 2.0*np.pi*x)
yf = scipy.fftpack.fft(y)
xf = np.linspace(0.0, 1.0/(2.0*T), N/2)

fig, ax = plt.subplots()
ax.plot(xf, 2.0/N * np.abs(yf[:N//2]))
plt.show()
```

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
N = 1024  #N
T =8*np.pi #T
t = np.linspace(0,T,N) #t
f = np.sin(17*t) + 2*np.sin(-97*t) + np.exp(-1j*57*t) #f(t)
F = np.fft.fft(f) #FFT
tf = np.linspace(0,N*np.pi/T,N//2 + 1) # \omega ,
plt.plot(tf, np.abs(F[:N//2+1])) # \omega ,  F . (N/2+1 )
plt.xlabel("$\omega$")
plt.ylabel("$F(\omega)$")
plt.title("The Results of FFT")
plt.show()
```
