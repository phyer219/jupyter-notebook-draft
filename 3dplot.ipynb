```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)

X = np.linspace(-1,1,100)
Y = np.linspace(-1,1,100)
X, Y = np.meshgrid(X, Y)
Z = np.sin(X)+ np.cos(Y)

ax.plot_surface(X, Y, Z,cmap=plt.get_cmap('hot'))
plt.show()
```
