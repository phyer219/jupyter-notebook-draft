```python
import numpy as np
from matplotlib import pyplot as plt

N = 1000

E = np.linspace(-5,5,N)
V = np.linspace(5, 50,N)
E, V = np.meshgrid(E, V)
L = np.sqrt(-E)
R = -np.sqrt(E+V)/np.tan(np.sqrt(E+V))
plt.contour(V, E,(L-R),[0])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 1/x

t1 = np.arange(-4.0, 4.0, 1)
t2 = np.arange(-4.0, 4.0, 0.02)

figure = plt.figure(1)

ax = figure.add_subplot(111)
ax.tick_params(labeltop = True)
ax.set_ylim([-40,40])

plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')
```
