```python
from matplotlib import pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
n = 100

fig = plt.figure()
ax = Axes3D(fig)
X = np.linspace(0, 5*np.pi,100 )
Y = np.linspace(0, 5*np.pi, 100)
X, Y = np.meshgrid(X, Y)
Z = np.sin(X)+np.sin(Y)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='rainbow')
```

```python
import numpy as np
hbar = 1.05e-34
a0 = 5.29e-11
delta = 0.1e-4
m = 6e-3/6.02e23
abg = 62*a0
mu = 9.27e-24
lam = np.sqrt((8*np.pi*hbar**2*abg*delta*mu)/(m))
print(lam)

reff = 4*np.pi*hbar**4/(m**2*lam**2)
print(reff/a0)

kf = 8.4e6
print(reff*kf)
```
